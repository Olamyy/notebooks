```
import gzip
data = []
with gzip.open('../content/gdrive/My Drive/colab_datasets/goodreads_books_poetry.json.gz') as f:
    for l in f:
        data.append(eval(l))
```

```
import pandas
```

```
df = pandas.DataFrame(data)
```

```
df.head()
```

```
df.columns
```

```
def drop_column(cols):
  return df.drop(cols, axis=1)

df = drop_column(['asin', 'authors', 'description', 'image_url', 'url', 'work_id', 'isbn', 'isbn13', 'kindle_asin', 'country_code', 'link'])
```

```
from sklearn.preprocessing import LabelEncoder
```

```
df.head()
```

```
le = LabelEncoder()

cat_columns = ['edition_information', 'format', 'is_ebook', 'language_code', 'publisher', 'publication_day', 'publication_month', 'publication_year', ]

for i in cat_columns:
  df[i] = le.fit_transform(df[i])
```
